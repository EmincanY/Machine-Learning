We will be using the same dataset through our discussions on classification with tree-methods (Decision Tree,Random Forests, and Gradient Boosted Trees) in order to compare performance metrics across these related models.

Gorman KB, Williams TD, Fraser WR (2014) Ecological Sexual Dimorphism and Environmental Variability within a Community of Antarctic Penguins (Genus Pygoscelis). PLoS ONE 9(3): e90081. doi:10.1371/journal.pone.0090081

Summary:
The data folder contains two CSV files. For intro courses/examples, you probably want to use the first one (penguins_size.csv).

* penguins_size.csv: Simplified data from original penguin data sets. Contains variables:

    * species: penguin species (Chinstrap, Adlie, or Gentoo)
    * culmen_length_mm: culmen length (mm)
    * culmen_depth_mm: culmen depth (mm)
    * flipper_length_mm: flipper length (mm)
    * body_mass_g: body mass (g)
    * island: island name (Dream, Torgersen, or Biscoe) in the Palmer Archipelago (Antarctica)
    * sex: penguin sex

* (Not used) penguins_lter.csv: Original combined data for 3 penguin species  

Note: The culmen is "the upper ridge of a bird's beak" 

**Our goal is to create a model that can help predict a species of a penguin based on physical attributes, then we can use that model to help researchers classify penguins in the field, instead of needing an experienced biologist**

Ordinal Encoder can be used with whole features at the same time, while labelEncoder can be used for at single feature.

## Some explanations about Label/Ordinal Encoder vs Dummies/OneHotEncoder

While extra features are created and model complexity increases with Dummies/OneHotEncoder, this does not happen in OrdinalEncoder or LabelEncoder.

In tree-based models, Dummies/OneHotEncoder can be misleading. OrdinalEncoder should be preferred.

In tree-based models, OrdinalEncoder is also more successful than Dummies/OneHotEncoder in terms of cost, speed and score.

Although Ordinal/Label Encoder gives good scores in tree-based models, dummies or OneHotEncoder gives better results in other models.

## Modelling and Model Performance

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix

def eval_metric(model, X_train, y_train, X_test, y_test):
    
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    
    print()
    
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
help(RandomForestClassifier)
```

1- N_ESTIMATORS -----> "How many tree models should I have in my model?" It is the most important parameter for Random Forest. (Default=100) creates 2/3 of the data with sub sample
!! The person who found Random Forrest (Leo Breiman) In his researches, it is said that 64 and 128 numbers must be tried for n_estimators.

2- CRITERION -----> GINI : It tries to minimize the value of the wrong calculation in the region it is in. So if the day is small, the mistake is also small. As the success increases, the day gets closer to 0. At the point where it separates all classes perfectly, gini=0. When Gini asks the distinguishing question in each branch, it chooses the class that gives the lowest value (ie, the least error). ENTROPY: It means chaos. We want chaos to be minimal like gini.
!!! As the success rate increases, the output and entropy approach 0. !!!

3- MAX_DEPTH------> How deep should the tree go? (Number of steps of arrows) (Adding bias)

4- MIN_SAMPLES_SPLIT ----> The number of samples that must be in that leaf in order for a leaf to be split. Default value=2. That is, "keep dividing if the number of samples is at least 2". If there is an error, it continues to divide, if not, it stops.

5- MAX_FEATURES -----> By browsing the random feature as much as we entered this parameter, "Which feature can I ask the best question?" says. This parameter needs to be set nicely to create different Decision Tree models. :
(Default='sqrt')

6- BOOTSTRAP ----> This parameter makes my model a Random Tree Model. It randomly separates the number of samples in the data at the rate of 2/3 for each tree and creates trees by selecting subsamples from them according to max_feature. Since it makes a random selection, it makes better generalizations in the train set and scores improve. (Default=True) !!

7- CLASS_WEIGHT -----> If desired, a balance operation like LogisticRegression can be applied. (Default=None)

8- MAX_SAMPLES -----> When we say 'none', it takes the data in all the data, but when we say 0.8, it takes 80% of the data. In other words, it gives 80% of the data when sampling each tree. (Default=None)

## Understanding Hyperparameters

```python
from sklearn.tree import plot_tree

def report_model(model):
    model_pred = model.predict(X_test)
    model_train_pred = model.predict(X_train)
    print('\n')
    print("Test Set")
    print(confusion_matrix(y_test, model_pred))
    print('\n')
    print(classification_report(y_test,model_pred))
    print('\n')
    print("Train Set")
    print(confusion_matrix(y_train, model_train_pred))
    print('\n')
    print(classification_report(y_train,model_train_pred))
    
    plt.figure(figsize=(12,8),dpi=100)
    plot_tree(model.estimators_[2], feature_names=X.columns,           # Our tree-based models report function.
          class_names=df.species.unique(),
          filled = True,
          fontsize = 8);
```

```python
pruned_tree = RandomForestClassifier()
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```
