```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

## Simple Linear Regression

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

```python
df['total_spend'] = df.TV + df.radio + df.newspaper # We do feature extraction because we want only 1 feature for simple linear regression
```

```python
df.head()
```

```python
sns.scatterplot(x = df['total_spend'] , y = df['sales']);
```

```python
sns.regplot(x = df['total_spend'] , y = df['sales'] ); # This is the best fit line. We want to create this.
```

```python
# Lets create that best fit line manually.
```

```python
X = df['total_spend']
y = df['sales']
```

```python
# y = ax + b
# y = mx + b
# y = B1x + B0
# help(np.polyfit)
```

```python
coefs = np.polyfit(X,y, deg = 1) # This is our coefficients.
coefs
```

```python
potential_spends = np.linspace(0,500 , 100)
```

```python
predicted_sales = potential_spends * coefs[0] + coefs[1]
predicted_sales
```

```python
plt.plot(potential_spends , predicted_sales);
```

```python
plt.figure(figsize=(10,6), dpi = 150)

sns.scatterplot(data = df , x = 'total_spend' , y = 'sales' )
plt.plot(potential_spends , predicted_sales , color = 'red'); # Same with sns.regplot
```

```python
spend = 200

predict = 200 * coefs[0] + coefs[1]

predict # Models say that if you spend 200 , you will have 13.98 sales.
```

```python
coefs3 = np.polyfit(X,y , deg = 3) # You can create polynominal equation too, if you want.

# y = B0 * x**3 + B1 * x**2 + B2 * x + B3

coefs3
```

```python
pot_spends = np.linspace(0,500,100)
```

```python
pred_sales = coefs3[0] * pot_spends**3 + coefs3[1] * pot_spends**2 + coefs3[2] * pot_spends + coefs3[3]
pred_sales
```

```python
plt.figure(figsize=(10,6), dpi = 150)

sns.scatterplot(data = df , x = 'total_spend' , y = 'sales' )
plt.plot(pot_spends , pred_sales , color = 'red'); # Degree 3. Simple Polynominal.
```
