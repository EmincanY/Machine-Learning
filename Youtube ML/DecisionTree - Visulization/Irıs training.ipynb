```python
from sklearn.datasets import load_iris
```

```python
df = load_iris()
```

```python
import numpy as np
```

```python
test_idx = [0,50,100]
```

```python
y_train = np.delete(df.target , test_idx)
X_train = np.delete(df.data , test_idx , axis = 0) 
```

```python
y_test = df.target[test_idx]
X_test = df.data[test_idx]
```

```python
from sklearn import tree
# from sklearn.tree import DecisionTreeClassifier
```

```python
cls = tree.DecisionTreeClassifier()
```

```python
cls.fit(X_train , y_train)
```

```python
y_test
```

```python
cls.predict(X_test)
```

# DecisionTree Visualizer

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,8) , dpi = 100)

tree.plot_tree(cls,
               feature_names=df.feature_names,
               rounded= True,
               filled=True,
               class_names = df.target_names,
               impurity=True
               )
plt.savefig('Irıs-train.png')
```
