```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder , StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("mushrooms.csv")
df.shape
```

```python
X = df.drop(columns='class')
y = df['class']

X_encoded = pd.get_dummies(X , prefix_sep = '__')
y_encoded = LabelEncoder().fit_transform(y)

X_encoded['Class'] = y_encoded
```

# Correlation Matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X_encoded.corr()['Class'].abs().nlargest(15)
```

```python
X_encoded.corr()['Class'].abs().nlargest(10).plot(kind='barh');
```

```python
X_highest_corr = X_encoded.corr()['Class'].abs().nlargest(10).index
```

```python
X_encoded[X_highest_corr]
```

```python
plt.figure(figsize = (12,8) , dpi = 100)

sns.heatmap(X_encoded[X_highest_corr].corr().abs() , annot=True);
```
