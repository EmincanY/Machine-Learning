```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer
```

```python
cancer.keys()
```

```python
print(cancer["DESCR"])
```

```python
X_train , X_test , y_train , y_test = train_test_split(cancer["data"],cancer["target"] , random_state = 0)
```

```python
train_correctness = []
test_correctness = []
```

```python
num_neighbors = range(1,11)
```

```python
for n_neigh in num_neighbors:
    cls = KNeighborsClassifier(n_neighbors=n_neigh)
    cls.fit(X_train,y_train)
    train_correctness.append(cls.score(X_train,y_train))
    test_correctness.append(cls.score(X_test,y_test))

plt.plot(num_neighbors, train_correctness, label = 'Train Correctness')
plt.plot(num_neighbors, test_correctness, label = 'Test Correctness')
plt.ylabel("Correctness")
plt.xlabel("n-Neighbors")
plt.legend()
plt.show()
```
