```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(iris["data"],iris["target"] , random_state=0 ) # this function get 2 parametere X and Y
```

```python
df_iris = pd.DataFrame(X_train, columns=iris.feature_names)
```

```python
scatter_matrix(df_iris,c = y_train , figsize=(15,15) , marker = "o" , hist_kwds={"bins":20},
               s=80,alpha = 0.8)
```

```python
# We will create K-NearestNeighbors Algorithm
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1) # We created KNN object.
```

```python
knn.fit(X_train,y_train) # We teach our model with fit.
```

```python
import numpy as np
X_new = np.array([[5,2.9,1,0.2]]) # Scikitlearn want 2 dimensions array to predict.
X_new.shape
```

```python
preds = knn.predict(X_new)
print(f"pred class : {preds}")
print(f"iris type :" , iris["target_names"][preds])
```

```python
y_preds = knn.predict(X_test)
print(y_preds)
print(y_test)
```

```python
print(np.mean(y_preds == y_test)) # mean() method for look properties of correctness.
```

```python
print(knn.score(X_test,y_test)) # This is also same with np.mean(). But this is easier
```
